```python
import pandas as pd
pd.set_option('display.max_colwidth',500)
data = pd.read_csv('without Labelling.csv',sep='\t',header=None)
data.columns =['review']
data
```

# Tokenizing

```python
from nltk.tokenize import word_tokenize
def tokenize(txt):
    tokens = word_tokenize(txt)
    return tokens

data['msg_clean_tokenized'] = data['review'].apply(lambda x: tokenize(x.lower()))
data
```

# POS Tagging

```python
import nltk
def tagging(tokens):
    pos = nltk.pos_tag(tokens)
    return pos
data['POS_Tags'] = data['msg_clean_tokenized'].apply(lambda x: tagging(x))
data
```

# extract only adjective

```python
import pandas as pd
from textblob import TextBlob

df=pd.read_csv('without labelling.csv')
df.columns =['reviews']

def pos_tag(text):
    try:
        return TextBlob(text).tags
    except:
        return None

df['pos'] = df['reviews'].apply(pos_tag)

```

```python
def get_adjectives(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "JJ"]

df['adjectives'] = df['reviews'].apply(get_adjectives)
df
```

```python
del df['reviews']
df
```

```python
del df['pos']
df
```

```python
df.to_csv('adjective.csv', index=False)
```
